```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_excel('file.xlsx', sheet_name='700MHz Availability')
df.drop(index=0, inplace=True)
```

```python
df.head()
df.tail()
```

```python
below_avaibility = df[df['Availability (%)'].between(99.5, 99.99)]
print(f"{below_avaibility['Site'].count()}")
below_avaibility.head()
```

```python
below_target= df[df['Availability (%)'] < 99.5]
below_target['Site'].count()
```

```python
below_avaibility.head()
```

```python
x = below_target['Availability (%)'].min()
plt.figure(figsize=(9,7))
plt.ylim((x-1),100, 0.4)
plt.axhline(y=99.5, color='red', linestyle='-')
plt.bar(below_avaibility.Site ,below_avaibility['Availability (%)'], color='g')
plt.bar(below_target.Site ,below_target['Availability (%)'], color='r')
plt.title(label="Febuary Sites Below 99.99")

plt.legend(labels=['Target', 'Feb avaibility'])
```
